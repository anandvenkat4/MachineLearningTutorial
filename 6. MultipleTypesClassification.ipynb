```python
import pandas as pd

data = pd.read_csv("datasets/german_credit_data_dataset.csv")
data.head()
```

```python
data.shape, data.columns
```

```python
data = data.drop(['telephone', 'personal', 'present_residence', 'other_installment_plans'], axis=1)
```

```python
data['savings'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

savings_dict = {"A65": 0, "A61": 1, "A62": 2, "A63": 3, "A64": 4}
data['savings'].replace(savings_dict, inplace=True)
```

```python
data = pd.get_dummies(data, columns=['checking_account_status',
                                    'credit_history',
                                    'purpose',
                                    'present_employment',
                                    'property',
                                    'housing',
                                    'other_debtors',
                                    'job',
                                    'foreign_worker'])

data.shape
data.head()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
def naive_bayes(x_train, y_train):
    
    clf = GaussianNB()
    clf.fit(x_train, y_train)
    
    return clf
```

```python
def k_nearest_neighbours(x_train, y_train):
    
    clf = KNeighborsClassifier(n_neighbors=10)
    clf.fit(x_train, y_train)
    
    return clf
```

```python
def svm(x_train, y_train):
    
    clf = SVC(kernel='rbf', gamma='scale')
    clf.fit(x_train, y_train)
    
    return clf
```

```python
def decision_trees(x_train, y_train):
    
    clf = DecisionTreeClassifier(max_depth=6)
    clf.fit(x_train, y_train)
    
    return clf
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score

X = data.drop('customer_type', axis=1)
Y = data['customer_type']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
def build_and_train_classfier(x_train, y_train, clf_fn):
    clf_model = clf_fn(x_train, y_train)
    y_pred = clf_model.predict(x_test)
    
    training_score = clf_model.score(x_train, y_train)
    test_score = accuracy_score(y_test, y_pred)
    prec_score = precision_score(y_test, y_pred)
    rec_score = recall_score(y_test, y_pred)
    
    print("Training Score: {0}, Test Score: {1}, Precision: {2}, Recall: {3}".format(training_score, test_score, prec_score, rec_score))
```

```python
build_and_train_classfier(x_train, y_train, naive_bayes)
build_and_train_classfier(x_train, y_train, k_nearest_neighbours)
build_and_train_classfier(x_train, y_train, svm)
build_and_train_classfier(x_train, y_train, decision_trees)
```
